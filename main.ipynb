```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np
# 设置中文字体为宋体
plt.rcParams['font.sans-serif'] = ['SimSun']
# 设置英文字体为新罗马
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rc('axes', unicode_minus=False) # 解决坐标轴负号显示问题
```

```python
raw_data = pd.read_excel('./算法工程师 工程实例2源数据.xls')
print(raw_data.info())
print(raw_data.isna().sum())
raw_data = raw_data.dropna()
plt.figure(figsize=(24,8))
plt.scatter(raw_data['A'],raw_data['B'])        #有过大异常值出现
plt.savefig('./fig/fig1.jpg')
raw_data =raw_data[raw_data['B']<5000] 
plt.figure(figsize=(24,8))
plt.scatter(raw_data['A'],raw_data['B'])        #有过大异常值出现
plt.savefig('./fig/fig1.jpg')
print(raw_data.info())
print(raw_data.isna().sum())
```

```python

# 选择某个时间段的数据，比如从 '2023-01-01' 到 '2023-03-31'
start_date = '2021-04-01'
end_date = '2021-04-07'
# 使用条件筛选
filtered_df = raw_data[(raw_data['A'] >= start_date) & (raw_data['A'] <= end_date)]
plt.figure(figsize=(24,8))
plt.plot(filtered_df['A'],filtered_df['D'])        #可以得出按天的规律性很强
plt.savefig('./fig/fig1.jpg')
```
